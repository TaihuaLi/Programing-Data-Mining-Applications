##### 5. Use Min-Max Normalization to transform the values of all numeric attributes in the original table (income, age, children) onto the range 0.0-1.0.

```python
dat_numeric = dat[['income', 'age', 'children']]
min_max_norm = lambda x: (x - min(x)) / (max(x)-min(x))
norm_numeric = dat_numeric.apply(min_max_norm)
norm_numeric.describe()
```
